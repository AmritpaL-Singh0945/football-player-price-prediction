
### Data Quality Issues Identified

1.Inconsistent Units (Height/Weight): The Height column mixes "cm" with "feet/inches" (e.g., "5'7" vs "170cm"), and Weight mixes "kg" with "lbs".

2.Currency Formatting (Value, Wage, Release Clause): Financial columns are stored as object strings with currency symbols ('€') and suffixes ('K', 'M') instead of numeric values (e.g., "€100M").

3.Embedded Characters: Columns like W/F, SM, and IR contain unnecessary special characters (e.g., "4 ★") preventing numerical analysis.

4.Hidden Newlines: The Club column contains hidden newline characters (\n) at the start of string values.

5.Incorrect Data Types: The Joined column is stored as a string rather than a datetime object, making time-based calculations impossible.

6.Mixed Data Types: Column 76 (Loan Date End) contains mixed types, causing memory warnings during load.

7.Redundant Columns: The Contract column duplicates information available in Joined and Contract Valid Until, and requires parsing to be useful.

```python
# sm is skill moves and ir is international rating 

def convert_height(height):
    if "cm" in height:
        return int(height.replace("cm",""))
    elif "'" in height:
        # Split 5'7" into [5, 7]
        feet,inches = height.split("'")
        inches = inches.replace("\"", "") 

        # Math: (Feet * 30.48) + (Inches * 2.54)
        return round((int(feet) * 30.48) + (int(inches) * 2.54))
    else:
        return height

# 2. Apply it to the Height column
df['Height'] = df['Height'].apply(convert_height)

# 3. Rename column to be specific
df.rename(columns={'Height': 'Height_cm'}, inplace=True)

# 4. Verify
print(df['Height_cm'].head())
print(df['Height_cm'].unique())
```

```python
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.replace("kg", ""))
    elif "lbs" in weight:
        lbs = int(weight.replace("lbs", ""))
        return round(lbs * 0.453592) # Standard conversion
    else:
        return weight


df['Weight'] = df['Weight'].apply(convert_weight)


df.rename(columns={'Weight': 'Weight_kg'}, inplace=True)

# 4. Verify
print(df['Weight_kg'].head())
print(df['Weight_kg'].unique())
```

```python
def convert_money(value):

    if pd.isna(value):
        return 0

    value = value.replace('€', '').strip()

    if 'M' in value:
        value = value.replace('M','')
        return float(value) * 1_000_000

    if 'K' in value:
        value = value.replace('K','')
        return float(value) * 1_000

    else:
        return float(value)
```

```python
cols_to_clean = ['Value', 'Wage', 'Release Clause']

for col in cols_to_clean:
    df[col] = df[col].apply(convert_money)

# Check the results.
print(df[cols_to_clean].head())
print("New Data Types:")
print(df[cols_to_clean].dtypes)
```

```python
#Redundant Columns: The Contract column duplicates information available in Joined and Contract Valid Until, 
#and requires parsing to be useful.

# 1. Define a function to parse the text
def extract_contract_info(contract):
    # Handle empty or 'Free' contracts
    if pd.isna(contract) or contract == 'Free':
        return np.nan, np.nan
    
    # Handle Loan players
    if 'On Loan' in contract:
        return np.nan, np.nan
    
    # Handle Standard "2018 ~ 2024" format
    if '~' in contract:
        start, end = contract.split(' ~ ')
        return int(start), int(end)
    
    # Default fallback
    return np.nan, np.nan

# 2. Apply the function
df['Contract_Start'], df['Contract_End'] = zip(*df['Contract'].apply(extract_contract_info))

# 3. Verify the results
# Look at the original column vs. the new columns
print(df[['Contract', 'Contract_Start', 'Contract_End']].head())

# 4. Check data types (Should be float because NaNs force integers to become floats)
print(df[['Contract_Start', 'Contract_End']].dtypes)
```

```python
# 1. Clean the 'Hits' column
def clean_hits(hits):
    if pd.isna(hits):
        return 0 # Assume 0 hits if missing
    
    hits = str(hits) # Ensure string
    
    if 'K' in hits:
        hits = hits.replace('K', '')
        return float(hits) * 1000
    else:
        return float(hits)

df['Hits'] = df['Hits'].apply(clean_hits)

# 2. Drop the mostly empty 'Loan Date End' column
df = df.drop(columns=['Loan Date End'])
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold

def target_encode_oof(df, col, target, n_splits=5, alpha=10):
    df = df.copy()
    
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
    global_mean = df[target].mean()
    
    encoded = np.zeros(len(df))
    
    for train_idx, val_idx in kf.split(df):
        train_df = df.iloc[train_idx]
        val_df = df.iloc[val_idx]
        
        stats = train_df.groupby(col)[target].agg(['mean', 'count'])
        
        smooth = (stats['mean'] * stats['count'] + global_mean * alpha) / (stats['count'] + alpha)
        
        encoded[val_idx] = val_df[col].map(smooth)
    
    # handle unseen categories
    encoded = np.where(pd.isnull(encoded), global_mean, encoded)
    
    return encoded
```

```python
def filter_features(df, target, corr_threshold=0.95, var_threshold=0.01):
    
    df = df.copy()
    
    print("Initial columns:", df.shape[1])
    
    X = df.drop(columns=[target])
    y = df[target]
    
    X = X.select_dtypes(include=[np.number])
    
    original_cols = set(X.columns)
    
    # -----------------------------
    # 🔹 Variance Filter
    # -----------------------------
    from sklearn.feature_selection import VarianceThreshold
    
    selector = VarianceThreshold(threshold=var_threshold)
    X_var = selector.fit_transform(X)
    
    var_cols = X.columns[selector.get_support()]
    X = pd.DataFrame(X_var, columns=var_cols)
    
    # -----------------------------
    # 🔹 Correlation Filter
    # -----------------------------
    corr_matrix = X.corr().abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    
    corr_drop = [col for col in upper.columns if any(upper[col] > corr_threshold)]
    X = X.drop(columns=corr_drop)
    
    # -----------------------------
    # 🔹 Target Correlation Filter
    # -----------------------------
    corr_with_target = X.apply(lambda col: col.corr(y))
    low_corr_cols = corr_with_target[abs(corr_with_target) < 0.01].index.tolist()
    
    X = X.drop(columns=low_corr_cols)
    
    # -----------------------------
    # 🔹 FINAL
    # -----------------------------
    final_cols = set(X.columns)
    
    removed_cols = list(original_cols - final_cols)
    
    print("Final columns:", len(final_cols))
    print("Removed columns:", len(removed_cols))
    
    return pd.concat([X, y], axis=1), removed_cols
    
```

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
import numpy as np

kf = KFold(n_splits=5, shuffle=True, random_state=42)

scores = []

for train_idx, val_idx in kf.split(X):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
    
    # Polynomial expansion
    poly = PolynomialFeatures(degree=2, include_bias=False)
    X_train_poly = poly.fit_transform(X_train)
    X_val_poly = poly.transform(X_val)
    
    # Scaling
    scaler = StandardScaler()
    X_train_poly = scaler.fit_transform(X_train_poly)
    X_val_poly = scaler.transform(X_val_poly)
    
    # Model
    model = LinearRegression()
    model.fit(X_train_poly, y_train)
    
    y_pred = model.predict(X_val_poly)
    scores.append(r2_score(y_val, y_pred))

print("Polynomial Regression (degree=2) R2:", np.mean(scores))
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
import numpy as np

kf = KFold(n_splits=5, shuffle=True, random_state=42)

scores = []

for train_idx, val_idx in kf.split(X):
    X_train, X_val = X.iloc[train_idx], X.iloc[val_idx]
    y_train, y_val = y.iloc[train_idx], y.iloc[val_idx]
    
    # Scaling (VERY IMPORTANT)
    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_val = scaler.transform(X_val)
    
    # SGD Model
    model = SGDRegressor(
        max_iter=1000,
        tol=1e-3,
        penalty='l2',   # Ridge-like
        alpha=0.0001,
        random_state=42
    )
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    
    scores.append(r2_score(y_val, y_pred))

print("SGD Regressor R2:", np.mean(scores))
```

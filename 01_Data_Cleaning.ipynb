### Data Quality Issues Identified

1.Inconsistent Units (Height/Weight): The Height column mixes "cm" with "feet/inches" (e.g., "5'7" vs "170cm"), and Weight mixes "kg" with "lbs".

2.Currency Formatting (Value, Wage, Release Clause): Financial columns are stored as object strings with currency symbols ('€') and suffixes ('K', 'M') instead of numeric values (e.g., "€100M").

3.Embedded Characters: Columns like W/F, SM, and IR contain unnecessary special characters (e.g., "4 ★") preventing numerical analysis.

4.Hidden Newlines: The Club column contains hidden newline characters (\n) at the start of string values.

5.Incorrect Data Types: The Joined column is stored as a string rather than a datetime object, making time-based calculations impossible.

6.Mixed Data Types: Column 76 (Loan Date End) contains mixed types, causing memory warnings during load.

7.Redundant Columns: The Contract column duplicates information available in Joined and Contract Valid Until, and requires parsing to be useful.

```python
def convert_height(height):
    if "cm" in height:
        return int(height.replace("cm",""))
    elif "'" in height:
        # Split 5'7" into [5, 7]
        feet,inches = height.split("'")
        inches = inches.replace("\"", "") 

        # Math: (Feet * 30.48) + (Inches * 2.54)
        return round((int(feet) * 30.48) + (int(inches) * 2.54))
    else:
        return height

# 2. Apply it to the Height column
df['Height'] = df['Height'].apply(convert_height)

# 3. Rename column to be specific
df.rename(columns={'Height': 'Height_cm'}, inplace=True)

# 4. Verify
print(df['Height_cm'].head())
print(df['Height_cm'].unique())
```

```python
def convert_weight(weight):
    if "kg" in weight:
        return int(weight.replace("kg", ""))
    elif "lbs" in weight:
        lbs = int(weight.replace("lbs", ""))
        return round(lbs * 0.453592) # Standard conversion
    else:
        return weight


df['Weight'] = df['Weight'].apply(convert_weight)


df.rename(columns={'Weight': 'Weight_kg'}, inplace=True)

# 4. Verify
print(df['Weight_kg'].head())
print(df['Weight_kg'].unique())
```

```python
def convert_money(value):

    if pd.isna(value):
        return 0

    value = value.replace('€', '').strip()

    if 'M' in value:
        value = value.replace('M','')
        return float(value) * 1_000_000

    if 'K' in value:
        value = value.replace('K','')
        return float(value) * 1_000

    else:
        return float(value)
```

```python
cols_to_clean = ['Value', 'Wage', 'Release Clause']

for col in cols_to_clean:
    df[col] = df[col].apply(convert_money)

# Check the results.
print(df[cols_to_clean].head())
print("New Data Types:")
print(df[cols_to_clean].dtypes)
```

```python
print(df['W/F'].unique())

star_cols = ['W/F', 'SM', 'IR']

for col in star_cols:
    df[col] = df[col].astype(str).str.replace('★', '').str.strip().astype(int)

# 4. Verify the results
print(df[star_cols].head())
print("New Data Types:")
print(df[star_cols].dtypes)
```

```python
# 1. Check the mess (printing values helps see the \n)

print("Before Cleaning (First 5 Clubs):")
print(df['Club'].head())

# 2. Apply strip() to remove leading/trailing whitespace & newlines

df['Club'] = df['Club'].str.strip()

# 3. Verify

print("After Cleaning:")
print(df['Club'].head())
print("Unique Names:")
print(df['Club'].unique())
```

```python
# 1. Check the current data type
print(f"Current Type: {df['Joined'].dtype}")

# 2. Convert to Datetime
df['Joined'] = pd.to_datetime(df['Joined'])

# 3. Verify the change
print(f"New Type: {df['Joined'].dtype}")

# 4. Check the actual values to ensure they look like dates (YYYY-MM-DD)
print(df['Joined'].head())
```

```python
# 1. Convert to Datetime
df['Loan Date End'] = pd.to_datetime(df['Loan Date End'], errors='coerce')

# errors='coerce' is crucial here. It tells Pandas: "If you find something that isn't a date, just turn it into NaT (Not a Time) instead of crashing."

# 2. Verify the Change
print(df[df['Loan Date End'].notna()]['Loan Date End'].head())

# 3. Check the Data Type
print(f"New Type: {df['Loan Date End'].dtype}")
```

```python
#Redundant Columns: The Contract column duplicates information available in Joined and Contract Valid Until, 
#and requires parsing to be useful.

# 1. Define a function to parse the text
def extract_contract_info(contract):
    # Handle empty or 'Free' contracts
    if pd.isna(contract) or contract == 'Free':
        return np.nan, np.nan
    
    # Handle Loan players
    if 'On Loan' in contract:
        return np.nan, np.nan
    
    # Handle Standard "2018 ~ 2024" format
    if '~' in contract:
        start, end = contract.split(' ~ ')
        return int(start), int(end)
    
    # Default fallback
    return np.nan, np.nan

# 2. Apply the function

df['Contract_Start'], df['Contract_End'] = zip(*df['Contract'].apply(extract_contract_info))

# 3. Verify the results
# Look at the original column vs. the new columns

print(df[['Contract', 'Contract_Start', 'Contract_End']].head())

# 4. Check data types (Should be float because NaNs force integers to become floats)

print(df[['Contract_Start', 'Contract_End']].dtypes)
```

```python
# Drop the 'Contract' column
df = df.drop(columns=['Contract'])
```

```python
# 1. Clean the 'Hits' column
def clean_hits(hits):
    if pd.isna(hits):
        return 0 # Assume 0 hits if missing
    
    hits = str(hits) # Ensure string
    
    if 'K' in hits:
        hits = hits.replace('K', '')
        return float(hits) * 1000
    else:
        return float(hits)

df['Hits'] = df['Hits'].apply(clean_hits)

# 2. Drop the mostly empty 'Loan Date End' column

df = df.drop(columns=['Loan Date End'])
```

```python
# Drop Personal Info Columns (Privacy/Irrelevance)
# We don't need URLs or IDs for statistical analysis

cols_to_drop = ['photoUrl', 'playerUrl', 'ID', 'LongName']
df = df.drop(columns=cols_to_drop)

# Final Verify
print(df.info())
```

```python
from datetime import datetime

df['Joined'] = pd.to_datetime(df['Joined'], errors='coerce')

# Extract useful info

df['Joined_year'] = df['Joined'].dt.year

current_year = datetime.now().year

df['Years_at_club'] = current_year - df['Joined_year']

df = df.drop(columns=['Joined'])
```

### 🧹 Data Preprocessing (Handling Missing Values, Encoding, and Scaling)

After identifying data quality issues, it is essential to preprocess the dataset to make it suitable for Machine Learning models.
Handling Missing Values
Missing data can lead to errors or biased predictions if not addressed properly. Filling missing values ensures that the model can learn from a complete dataset without discarding useful information.
Encoding Categorical Variables
Machine Learning models cannot work with categorical data directly. Encoding transforms categorical features (such as player position or club) into numerical form, allowing the model to interpret and learn patterns from them.
Feature Scaling
Features in the dataset exist on different scales (e.g., wages in millions vs ratings out of 100). Scaling ensures that no single feature dominates the model due to its magnitude, leading to more stable and accurate predictions.
These steps help standardize the dataset, reduce bias, and improve overall model performance.
